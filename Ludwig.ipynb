```python
import requests

url = 'http://boston.lti.cs.cmu.edu/classes/95-865-K/HW/HW2/reuters-allcats-6.zip'
response = requests.get(url)

# Check for a valid response
if response.status_code == 200:
    # Save the content to a file
    with open('reuters-allcats-6.zip', 'wb') as file:
        file.write(response.content)
else:
    print(f'Failed to retrieve the file: {response.status_code}')
```

```python
import zipfile
import os

zip_path = '/content/reuters-allcats-6.zip'  # Path to your zip file
extract_path = '.'  # Path where you want to extract the files

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# Optional: List the files in the extraction path to verify
for root, dirs, files in os.walk(extract_path):
    for file in files:
        print(os.path.join(root, file))
```
